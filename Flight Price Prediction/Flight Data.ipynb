### Importing Required Library

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

import warnings
warnings.filterwarnings('ignore')
```

### Connect to the Web driver

```python
driver = webdriver.Chrome('chromedriver.exe')
```

### Open Url

```python
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destinationCountry_0=IN&destinationCountry_1=IN&destinationCountry_2=IN&destinationCountry_3=IN&destinationCountry_4=IN&destination_0=BOM&destination_1=GOI&destination_2=DEL&destination_3=JAI&destination_4=DEL&flexi=0&flight_depart_date_0=22%2F10%2F2021&flight_depart_date_1=26%2F10%2F2021&flight_depart_date_2=27%2F11%2F2021&flight_depart_date_3=10%2F01%2F2022&flight_depart_date_4=27%2F01%2F2022&hb=0&noOfSegments=5&originCountry_0=IN&originCountry_1=IN&originCountry_2=IN&originCountry_3=IN&originCountry_4=IN&origin_0=DEL&origin_1=BOM&origin_2=GOI&origin_3=DEL&origin_4=CCU&type=M&unique=739518569126&viewName=normal"
driver.get(url)
```

### List of data scraping

```python
airline = []
deptime = []
arrivaltime = []
deploc = []
arrivalloc = []
stop = []
duration = []
price = []
```

### Extract 1st page data

```python
# Extract Airline Name 
xpath = driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')
for i in xpath:
    try:
        airline.append(i.text)
    except:
        airline.append("-")
        
print('Length of Airline Name ------->',len(airline))

# Extract Departure Time
xpath = driver.find_elements_by_xpath('//div[@class="i-b pr"]')
for i in xpath:
    try:
        deptime.append(i.text)
    except:
        deptime.append("-")
        
print('Length of Airline Departure Time ------->',len(deptime))

# Extract Arrival Time
xpath = driver.find_elements_by_xpath('//p[@class="bold fs-15 mb-2 pr time"]')
for i in xpath:
    try:
        arrivaltime.append(i.text)
    except:
        arrivaltime.append("-")
        
print('Length of Airline Arrival Time ------->',len(arrivaltime))

# Extract Departure Location
xpath = driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
for i in xpath:
    try:
        deploc.append(i.text)
    except:
        deploc.append("-")
        
print('Length of Airline Departure Location ------->',len(deploc))

# Extract Arrival Location
xpath = driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
for i in xpath:
    try:
        arrivalloc.append(i.text)
    except:
        arrivalloc.append("-")
        
print('Length of Airline Arrival Location ------->',len(arrivalloc))

# Extract Airline Stop
xpath = driver.find_elements_by_xpath('//span[@class="dotted-borderbtm"]')
for i in xpath:
    try:
        stop.append(i.text)
    except:
        stop.append("-")
        
print('Length of Airline Stops ------->',len(stop))

# Extract Airline Duration
xpath = driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')
for i in xpath:
    try:
        duration.append(i.text)
    except:
        duration.append("-")
        
print('Length of Airline Duration ------->',len(duration))

# Extract Airline Price
xpath = driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
for i in xpath:
    try:
        price.append(i.text)
    except:
        price.append("-")
        
print('Length of Price ------->',len(price))
```
